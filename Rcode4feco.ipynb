```r
# Loading required libraries
library(tseries)   # For ADF and KPSS tests
library(rugarch)   # For GARCH modeling
library(forecast)  # For ARIMA modeling and ACF/PACF
library(ggplot2)   # For plotting
library(zoo)       # For na.locf

# Print working directory for debugging
cat("Working directory:", getwd(), "\n")

# Check if CSV file exists
csv_file <- "vn30_closing_prices.csv"
if (!file.exists(csv_file)) {
  stop("CSV file not found in working directory: ", getwd())
}

# Create directories for saving plots
plots_dir <- "plots"
acf_pacf_dir <- file.path(plots_dir, "acf_pacf")
if (!dir.exists(plots_dir)) {
  dir.create(plots_dir)
  cat("Created directory:", plots_dir, "\n")
}
if (!dir.exists(acf_pacf_dir)) {
  dir.create(acf_pacf_dir)
  cat("Created directory:", acf_pacf_dir, "\n")
}

# Reading the data
data <- read.csv(csv_file, stringsAsFactors = FALSE)

# Converting time column to Date
data$time <- as.Date(data$time)

# Selecting last 45 days for forecasting
n_forecast <- 45
n_total <- nrow(data)
train_end_idx <- n_total - n_forecast
train_data <- data[1:train_end_idx, ]
test_data <- data[(train_end_idx + 1):n_total, ]

# Getting list of stock columns (excluding 'time')
stock_columns <- setdiff(names(data), "time")

# Initializing lists and data frames
forecast_results <- list()
mae_results <- data.frame(Stock = character(), MAE = numeric(), stringsAsFactors = FALSE)
skipped_stocks <- data.frame(Stock = character(), Reason = character(), stringsAsFactors = FALSE)
correct_direction_counter <- 0
total_forecasted_stocks <- 0

# Minimum number of observations for modeling
min_obs <- 50

# Function to determine ARIMA orders
determine_arima_orders <- function(series, stock) {
  # Initialize differencing order
  d <- 0
  max_d <- 2
  adf_pvalue <- 1
  kpss_pvalue <- 0

  # Test for stationarity with ADF and KPSS
  while (d <= max_d) {
    diff_series <- if (d == 0) series else diff(series, differences = d)
    diff_series <- na.omit(diff_series)

    if (length(diff_series) < min_obs) {
      return(list(d = NA, p = NA, q = NA, reason = paste("Insufficient data after differencing (", length(diff_series), " observations)")))
    }

    # ADF test (p-value < 0.05 indicates stationarity)
    adf_result <- tryCatch({
      adf.test(diff_series, alternative = "stationary")
    }, error = function(e) {
      return(list(p.value = 1))
    })
    adf_pvalue <- adf_result$p.value

    # KPSS test (p-value > 0.05 indicates stationarity)
    kpss_result <- tryCatch({
      kpss.test(diff_series, null = "Level")
    }, error = function(e) {
      return(list(p.value = 0))
    })
    kpss_pvalue <- kpss_result$p.value

    if (adf_pvalue < 0.05 && kpss_pvalue > 0.05) {
      break
    }
    d <- d + 1
  }

  if (d > max_d) {
    return(list(d = NA, p = NA, q = NA, reason = "Series not stationary after max differencing"))
  }

  # ACF/PACF for p and q
  acf_vals <- acf(diff_series, plot = FALSE, lag.max = 20)$acf[-1]
  pacf_vals <- pacf(diff_series, plot = FALSE, lag.max = 20)$acf
  conf_int <- 1.96 / sqrt(length(diff_series))  # 95% confidence interval

  # Save ACF/PACF plots
  png(file.path(acf_pacf_dir, paste0(stock, "_acf_pacf.png")), width = 800, height = 400)
  par(mfrow = c(1, 2))
  acf(diff_series, lag.max = 20, main = paste("ACF for", stock, "(d =", d, ")"))
  pacf(diff_series, lag.max = 20, main = paste("PACF for", stock, "(d =", d, ")"))
  par(mfrow = c(1, 1))
  dev.off()

  # Determine p and q based on significant lags
  p <- sum(abs(pacf_vals) > conf_int)
  q <- sum(abs(acf_vals) > conf_int)
  p <- min(p, 5)  # Cap at 5 to avoid overfitting
  q <- min(q, 5)

  return(list(d = d, p = p, q = q, reason = NULL))
}

# Looping through each stock
for (stock in stock_columns) {
  cat("\nProcessing stock:", stock, "\n")

  # Extracting prices for the stock
  prices <- train_data[[stock]]

  # Imputing missing prices with forward fill
  prices <- na.locf(prices, na.rm = FALSE)

  # Handling remaining missing or non-positive prices
  if (any(is.na(prices))) {
    cat("Skipping", stock, "due to remaining NA prices after forward fill\n")
    skipped_stocks <- rbind(skipped_stocks, data.frame(Stock = stock, Reason = "Remaining NA prices after forward fill"))
    next
  }
  prices[prices <= 0] <- 0.01  # Replace non-positive with small positive

  # Calculating log-returns
  log_returns <- diff(log(prices))

  # Dropping NA values in log-returns
  log_returns <- na.omit(log_returns)

  # Checking if enough data remains
  if (length(log_returns) < min_obs) {
    cat("Skipping", stock, "due to insufficient data after dropping NAs (", length(log_returns), " observations)\n")
    skipped_stocks <- rbind(skipped_stocks, data.frame(Stock = stock, Reason = paste("Insufficient data (", length(log_returns), " observations)")))
    next
  }

  # Determining ARIMA orders
  arima_orders <- determine_arima_orders(log_returns, stock)
  if (!is.null(arima_orders$reason)) {
    cat("Skipping", stock, "due to", arima_orders$reason, "\n")
    skipped_stocks <- rbind(skipped_stocks, data.frame(Stock = stock, Reason = arima_orders$reason))
    next
  }

  d <- arima_orders$d
  p <- arima_orders$p
  q <- arima_orders$q
  cat("Selected ARIMA order for", stock, ": (", p, ",", d, ",", q, ")\n")

  # Fitting ARIMA/GARCH model
  tryCatch({
    # Specifying GARCH model with selected ARIMA order
    spec <- ugarchspec(
      variance.model = list(model = "sGARCH", garchOrder = c(1, 1)),
      mean.model = list(armaOrder = c(p, q), include.mean = TRUE),
      distribution.model = "norm"
    )

    # Fitting the model
    fit <- ugarchfit(spec = spec, data = log_returns)

    # Checking residuals
    residuals <- residuals(fit, standardize = TRUE)

    # Saving residual diagnostics plots
    png(file.path(plots_dir, paste0(stock, "_residuals.png")), width = 800, height = 400)
    par(mfrow = c(1, 2))
    qqnorm(residuals, main = paste("Q-Q Plot of Residuals for", stock))
    qqline(residuals, col = "red")
    acf(residuals, main = paste("ACF of Residuals for", stock))
    par(mfrow = c(1, 1))
    dev.off()
    cat("Saved residual plot for", stock, "to", file.path(plots_dir, paste0(stock, "_residuals.png")), "\n")

    # Ljung-Box test for residual autocorrelation
    lb_test <- Box.test(residuals, lag = 10, type = "Ljung-Box")
    cat("Ljung-Box Test p-value for", stock, ":", lb_test$p.value, "\n")

    # Forecasting log-returns for 45 days
    forecast <- ugarchforecast(fit, n.ahead = n_forecast)
    forecast_mean <- forecast@forecast$seriesFor
    forecast_sigma <- forecast@forecast$sigmaFor

    # Converting log-returns to prices
    last_price <- tail(train_data[[stock]], 1)
    cumulative_returns <- cumsum(forecast_mean)
    forecast_prices <- last_price * exp(cumulative_returns)

    # Preparing data for plotting
    dates_forecast <- test_data$time
    actual_prices <- test_data[[stock]]
    plot_data <- data.frame(
      Date = dates_forecast,
      Actual = actual_prices,
      Forecast = forecast_prices
    )

    # Calculating MAE
    mae <- mean(abs(actual_prices - forecast_prices))
    mae_results <- rbind(mae_results, data.frame(Stock = stock, MAE = mae))

    # Checking direction of forecast vs actual
    forecast_increase <- forecast_prices[1] < forecast_prices[n_forecast]
    actual_increase <- actual_prices[1] < actual_prices[n_forecast]
    if (forecast_increase == actual_increase) {
      correct_direction_counter <- correct_direction_counter + 1
      cat("Correct direction for", stock, ": Forecast and actual both",
          ifelse(forecast_increase, "increase", "decrease"), "\n")
    } else {
      cat("Incorrect direction for", stock, ": Forecast",
          ifelse(forecast_increase, "increase", "decrease"),
          ", Actual", ifelse(actual_increase, "increase", "decrease"), "\n")
    }

    # Increment total forecasted stocks
    total_forecasted_stocks <- total_forecasted_stocks + 1

    # Plotting actual vs forecasted prices
    p <- ggplot(plot_data, aes(x = as.Date(Date))) +
      geom_line(aes(y = Actual, color = "Actual"), size = 1) +
      geom_line(aes(y = Forecast, color = "Forecast"), size = 1, linetype = "dashed") +
      labs(title = paste(stock, "Stock Price: Actual vs Forecast"),
           x = "Date", y = "Price (VND)",
           color = "Series") +
      theme_minimal() +
      scale_color_manual(values = c("Actual" = "blue", "Forecast" = "red")) +
      scale_x_date(date_labels = "%Y-%m-%d", breaks = "7 days")  # Correct date format and breaks

    # Displaying the plot
    print(p)

    # Saving the plot
    ggsave(file.path(plots_dir, paste0(stock, "_forecast.png")), plot = p, width = 8, height = 4)
    cat("Saved forecast plot for", stock, "to", file.path(plots_dir, paste0(stock, "_forecast.png")), "\n")

    # Storing forecast results
    forecast_results[[stock]] <- plot_data

  }, error = function(e) {
    cat("Error fitting model for", stock, ":", conditionMessage(e), "\n")
    skipped_stocks <- rbind(skipped_stocks, data.frame(Stock = stock, Reason = paste("Model fitting error:", conditionMessage(e))))
  })
}

# Calculating success probability
success_probability <- if (total_forecasted_stocks > 0) {
  (correct_direction_counter / total_forecasted_stocks) * 100
} else {
  0
}

# Saving forecast results, MAE, and skipped stocks
saveRDS(forecast_results, "forecast_results.rds")
write.csv(mae_results, "mae_results.csv", row.names = FALSE)
write.csv(skipped_stocks, "skipped_stocks.csv", row.names = FALSE)

# Saving direction summary
direction_summary <- data.frame(
  Total_Forecasted_Stocks = total_forecasted_stocks,
  Correct_Direction_Counter = correct_direction_counter,
  Success_Probability_Percent = success_probability
)
write.csv(direction_summary, "direction_summary.csv", row.names = FALSE)

# Printing summaries
cat("\nForecasting complete.\n")
cat("Plots saved in:", file.path(getwd(), plots_dir), "\n")
cat("ACF/PACF plots saved in:", file.path(getwd(), acf_pacf_dir), "\n")
cat("MAE results saved in:", file.path(getwd(), "mae_results.csv"), "\n")
cat("Forecast data saved in:", file.path(getwd(), "forecast_results.rds"), "\n")
cat("Skipped stocks saved in:", file.path(getwd(), "skipped_stocks.csv"), "\n")
cat("Direction summary saved in:", file.path(getwd(), "direction_summary.csv"), "\n")

cat("\nMAE Summary:\n")
print(mae_results)

cat("\nSkipped Stocks Summary:\n")
if (nrow(skipped_stocks) > 0) {
  print(skipped_stocks)
} else {
  cat("No stocks were skipped.\n")
}

cat("\nDirection Summary:\n")
cat("Total Forecasted Stocks:", total_forecasted_stocks, "\n")
cat("Correct Direction Counter:", correct_direction_counter, "\n")
cat("Success Probability:", round(success_probability, 2), "%\n")
```
